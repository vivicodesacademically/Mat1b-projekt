```python
import dtumathtools
import numpy as np
 



def assemble_K_general(nodes, a_func=None):
    """Saml K for vilkårligt net og funktion a(x)."""
    N = len(nodes) - 2
    K = np.zeros((N, N))
    print(K)
    if a_func is None:
        a_func = lambda x: 1.0
    for e in range(N + 1):
        e_1= nodes[e]
        e_2= nodes[e+1]
        
        
        h_e = e_1 + e_2  # elementbredde
        x_mid = (e_1+e_2)/2  # midtpunkt af elementet
        a_e = a_func(x_mid)  # a(x) evalueret i midtpunktet
        i = e - 1
        j = e
        
        # opdater K[i,i], K[j,j], K[i,j], K[j,i]
        # Husk at tjekke om i og j er indre knudepunkter
        k=(a_e/h_e)*np.array([[1,-1],[-1,1]])
        if 0<=j<N:
            K[j,j]+=k[1,1]
        if 0<=i<N:
            K[i,i]+=k[0,0]
        if 0<=i<N and 0<=j<N:
            K[i,j]+=k[0,1]
            k[j,i]+=k[1,0]
        
        print(K,k)
    return K
```
